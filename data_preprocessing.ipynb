```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.neighbors import KDTree
```

```python
import pickle as pk
```

```python
df = pd.read_csv('../data/cdrsampledata.csv', header=None)
```

```python
df
```

```python
df.dropna(axis=0, subset=[5, 6], inplace=True)
```

```python
len(df[[5, 6]].drop_duplicates())
```

```python
bs_coordinates = df[[5, 6]].drop_duplicates().values
```

```python
kdtree = KDTree(bs_coordinates)
```

```python
df[8] = kdtree.query(df[[5, 6]].values, return_distance=False)[:, 0]
```

```python
df[1].max()
```

```python
df[1].min()
```

```python
df[9] = (pd.to_datetime(df[1]) - pd.to_datetime('2013-08-01 00:02:05')) // pd.Timedelta('00:30:00')
```

```python
df[10] = df[0].astype('category').cat.codes
```

```python
df[11] = df[9] // 48
df[12] = df[9] % 48
```
